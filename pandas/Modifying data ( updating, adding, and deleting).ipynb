#### Updating Columns

We'll start off by learning how to update columns then move on to rows.

So before we change anything, lets take a look at our columns using the `.columns` attribute 

```python
people_df.columns
```

We see we have a couple of columns (5 in total) now what if we want to update the `first` and `last` to be more specific as in `first_name` 

There are a couple of ways we can do that, one of them is use the same `.columns` attribute to re-assign them as such

```python
people_df.columns = ['first_name', 'last_name', 'email', 'year', 'hobby']
```

```python
people_df.columns
```

This approach is rarly used (only when you need to change all of your columns) so it's best not to use this approach.

A common use cause is when you want to change the columns say to upper or lowercase or remove spaces and of the sort. what you do in this case is use a list comprehension.

```python
people_df.columns = [x.upper() for x in people_df]
people_df
```

Say your columns have spaces in their names and you want to remove them, so you do something similar by 

```python
people_df.columns = people_df.columns.str.replace(' ', '_')
people_df
```

So far the examples we've used changes all the values. What if we wanted to change just one column? 
in this case we use the `.rename` method where we pass a dictionary of the columns we want to change.

__note__ that the key is the current column name you want to chane and the value is the new value for the column. Also make sure to persist the changes with `inplace=True`

```python
people_df.rename(columns={'first_name': 'first', 'last_name': 'last'}, inplace=True)
```

```python
people_df
```

That's all there is to it when it comes to columns, now we'll move on to updating data in the rows since there's so much that comes with it

#### Updating Rows
First things first, let's look at how to update a single value. Previouly we saw how we could look up values using `.loc` and `iloc` we'll be using that for our starting point to set our values as well.

So let's start by grabing the row of `john doe` and change his last name to `smith` 

```python
people_df.loc[(people_df['first']=='john') & (people_df['last']=='doe')]
```

There are a couple of ways to update the data
1. is by selecting the row (like we did in the cell above) and assigning it to an array of the values (they need to be all the values) `people_df.loc[2] = ['john','smith','john@email.com', 1981, 'sports']`
2. Now what if we had a lot of columns, it would be a pain to pass in all those data again so another way is to specify the columns using `loc` as seen in the cell bellow

```python
people_df.loc[2,['last', 'email']] = ['smith', 'johnsmith@email.com']
```

```python
people_df
```

Now pandas also has a `.at` to change a single value (though the instructor perfers the `.loc` option) lets take an example

```python
people_df.at[2, 'email'] = 'john@email.com'
people_df
```

A common mistak that people do is trying to chnage the value without using one of those indexers `.loc` or `.at` lets look at an example of changing a last name back to `doe`

```python
# filter the data
filt = (people_df['first'] == 'john') & (people_df['last']=='smith')
people_df[filt]['last'] = 'doe' # error
```

```python
people_df
```

As you could see we got an error and the value didn't change. so if we wanted to change it, we can use it exactly but with the `.loc` indexer as follows

```python
filt = (people_df['first'] == 'john') & (people_df['last']=='smith')
people_df.loc[filt, 'last'] = 'doe'
people_df
```

now that was for a single row of data, but what about multiple rows and same as everything, there are multiple ways to do so. 

For our first example, let's assume we want to change the `last` to uppercase we do that by addigning that column an uppercase of itself as such

```python
people_df['last'].str.upper()
```

```python
people_df
```

Now that just returned a series of the result it didn't chnage it. for it to chnage we have to assign it the new value 

```python
people_df['last'] = people_df['last'].str.upper()
people_df
```

Now what if we wanted to do something a bit more advanced. There are a couple of ways to do so and we'll be going over the 4 popular methods. Note that people do get them confused and rightfuly so given that they're very similar but we'll go over them and help clarify. 

Just to get an idea, the methods we'll be talking about are 
- `apply`
- `map`
- `applymap`
- `replace`

##### Method 1 - apply
`apply` is used for calling a function on our values and it can work on either a dataframe or a series object and the behaviour might be a little different than you expect for each of those different objects. 

Let's look at how apply works for a __series__ when used on it, it would apply a function on every value in our series. So say we want to see the length of our email address, maybe we have a website that has a limitation on the length of the email. We can apply the `len` function to each value in our series by doing something like this 

```python
people_df['email'].apply(len)
```

Now we used to check for data, but we can also use it to update vales as well. Let's make a simple function that returns an upper case of the passed in value. (note this is a super simple example that coule have been done as we did before with the uppercase list but we're showing an example of a custom fucntion and know it can be much more complicated

```python
def update_name(name):
    return name.upper()
```

```python
people_df['first'].apply(update_name)
```

__Note__ that this didn't change our values we need to assign it so

```python
people_df['first'] = people_df['first'].apply(update_name)
```

```python
people_df
```

For simple methods, people tend to use `lambda` functions which are basically anonymous functions without a specific name that we can use for things like this. 

let's take an example to see how it would work. let's revert them back to lower case. We'll pass in a `lambda` function to `apply` rather than a custom function

```python
people_df['first'] = people_df['first'].apply(lambda x: x.lower())
```

```python
people_df
```

Now we saw how `apply` works with __Series__ now it's time to look at how it works with __data frames__ now when we ran apply on a series it ran the function on every value in the list of values in the series. Now with data frames it will run it on every row or column of that dataframe. 

Let's take an example to clarify. taking the len example from before

```python
people_df['email'].apply(len) # series 
```

```python
people_df.apply(len) #data frame
```

So we didn't get what we expected. So what happened is that it ran the function on each series, specificly the column. So it counted the number of items in each column. 

So it's telling us that first has 6 values `['athoug', 'jane', 'john', 'sam', 'zack', 'olivia']` so that's the number of each row in a column

```python
len(people_df['first'])
```

__note__ that we can apply it to the rows by changing teh axes as follows

```python
people_df.apply(len, axis='columns')
```

```python
people_df.apply(pd.Series.min)
```

We can also use a `lambda` function as well but note that it works ona series object so the parameter (in this case `x`) is going to be a series, it won't be a value. so what methods does a series have, well it has a `min`

```python
people_df.apply(lambda x: x.min())
```

It's more useful if our dataframe includes numberical data for example we can use numpy to apply the square root for all our series objects or any type of numerical analysis like that.

so using apply 
- on a __series__ it applies the function of every value in the series 
- on a __data frame__ it applies the function on every series in teh data frame

Now what if we want to apply a function on every element in the data frame, and that's what `applymap` is used for

##### Mathod 2 - applymap
`applymap` only works on __data frames__ let's see how this is different from `apply`

Let's use the same `len` example with `applymap` and see what happenes

```python
people_df.applymap(len)
```

Since all our values are strings we can apply the `upper` method on all the values as follows

```python
people_df.applymap(str.lower)
```

##### Method 3 - map
`map` only works on a __series__ it is used to subsitute each value in a series with another value.

So let's say for example we want to subsitute a couple of our first names. to do so, we first access the column, then use the `map` method on them, and then pass in a dictionary with teh values we want to subsitute 

```python
people_df['first'].map({'olivia': 'love'})
```

```python
people_df
```

First we see that the changes isn't applied and we also see that `NaN` is applied, and we don't want that so if we want to change a value we need to use teh `replace` method

##### Mathod 4 - replace

```python
people_df['first'].replace({'olivia': 'love'})
```

```python
people_df
```

Now the changes didn't reflect because we need to assign it sd such

```python
people_df['first'] = people_df['first'].replace({'olivia': 'love'})
```

```python
people_df
```

### Adding and removing columns
Let's explore adding columns first because it's the easiest bit. so we can just create a column and pass in a series of values we want it to have. 

Let's use teh people's dataset for practice.

```python
people_df
```

Let's say we want to combine the `first` and `last` column and simply call it `full_name`

So first to get a series of the first and last combine we could simply say 

```python
people_df['first'] + ' ' + people_df['last']
```

```python
people_df['full_name'] = people_df['first'] + ' ' + people_df['last']
```

```python
people_df
```

__Note__ that you can't use the dot notation to assign a column, you have to use the brakets notation. Because if you try the dot notation then python would think you're trying to assign an attribute to the dataframe

Let's move on to removing columns. Now that we have our `full_name` column, we no longer need the `first` and `last` columns. so what we do is use the `drop` method on our __data frame__ it's as easy as

```python
people_df.drop(columns=['first', 'last'])
```

```python
people_df
```

Now was we can see it just shows us what our dataframe would look like if we want to keep the changed we have to use teh `inplace=True` argument with the `drop` method

```python
people_df.drop(columns=['first', 'last'], inplace=True)
people_df
```

now if we want to revert back to our original `first` and `last` what we do is use that string `split` method from before. So let's try that on our `full_name` column

```python
people_df['full_name'].str.split(' ', expand=True)
```

```python
people_df[['first', 'last']] = people_df['full_name'].str.split(' ', expand=True)
```

```python
people_df
```

We now explored adding and removing columns, now it's time to explore adding and removing rows of data.

There are a couple of ways to add rows to our data frame. 
1. First we might just want to add a single row to our data frame 
2. another way is if we wanted to combine two data frames together into a single data frame by appending the rows of one to another

#### Adding a single row of data
we can do this using the `append` method so if I want to add a single row, then I can simply do so by 

```python
people_df.append({'first': 'Ethan'}, ignore_index=True)
```

Let's update the data we'll be playig with

```python
people = {
    'first': ['Ethan', 'Hila',],
    'last': ['Klien', 'doe',],
    'email':['ethan@email.com', 'hila@email.com',],
}
df2 = pd.DataFrame(people)
df2
```

Now let's say we want to add this newly created dataframe to our already existing one. What we can do is, simply `append` the one dataframe to the other 

```python
people_df.append(df2, ignore_index=True)
```

```python
people_df
```

Now unlike the drop method we can't have it persist so the way to do it when adding a new dataframe to another is by assigning it the new append as so

```python
people_df = people_df.append(df2, ignore_index=True)
```

```python
people_df
```
