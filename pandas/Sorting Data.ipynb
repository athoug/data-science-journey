### Practicing
as we usually do, we first practice what we learn ona smaller dataset and then upgrade to the boheamoth dataset so let's start by creating the people dataframe

```python
people = {
    'first': ['athoug', 'jane', 'john', 'sam', 'zack', 'olivia', 'adam'],
    'last': ['alsoughayer', 'doe', 'doe', 'simples', 'something', 'mendez', 'doe'],
    'email':['athoug@email.com', 'jane@email.com', 'john@email.com', 'sam@email.com', 'zack@email.com', 'olivia@email.com', 'adam@email.com'],
    'year': [1990, 1978, 1981, 1997, 1994, 1999, 2000],
    'hobby': ['making', 'cooking', 'sports', 'crafts', 'music', 'drama', 'music']
}
```

```python
people_df = pd.DataFrame(people)
people_df
```

now looking at this dataset we can look at different ways to sort it. One way is to sort it based on `last` name, and we do that with the `sort_values` method and set the `by` argument to the column name we want. 

```python
people_df.sort_values(by='last')
```

Now we can see that it's alphabetically which is the default, now if we want it to be decsending then what we do is set the argument `ascending` to `False` 

```python
people_df.sort_values(by='last', ascending=False)
```

at times our sorts might be a bit more complicated where we want to sort on multiple values and we do this by say we have identical values in one of our columns so we use the second column to sort based on it. So lets take an example

say we want to sort `last` name in descending order and then if there are any duplicate values we do sort based on `first`. To do that, we pass in a list of columns to the `by` argument in `sort_values`

```python
people_df.sort_values(by=['last', 'first'], ascending=False)
```

Now what if you want to sort one column in `ascending` order and the other in `descending` order well to do that what we have to do is we simply have to passby a list of boolean values to the `ascending` argument

```python
people_df.sort_values(by=['last', 'first'], ascending=[False, True])
```

__Note__ that the dataframe in pandas doesn't carry on the changes we make as we say here and in other exercises. To make the changes permanint we need to set the `inplace=True` in the `sort_values` method

```python
people_df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
people_df
```

Now if I want to revert this and have them in the previous order what we do is we sort by index and to do that we use `sort_index` method (remember to keep the changes you must set the `inplace=True`

```python
people_df.sort_index()
```

```python
people_df
```

If you want to sort a single column it's good to pointout that you can sort a single series (as in that column) itself because they have a sort method as well. It uses the same method `sort_values`

```python
people_df['last'].sort_values()
```

One way we can sort the values is by county. Maybe we're doing analysis for people in different countries, so lets sort them and see how we can sort by `county` and `salary` 

```python
df.sort_values(by='Country', inplace=True)
```

Now lets take a look at the `country` column and view the first 50 results as well as the `salary`

```python
df[['Country', 'CompTotal']].head(50)
```

Now what we want to do is sort countries in `ascending` order and the salary in `descending` order so just like we practiced before what we have to do is pass in a list of booleans 

```python
df.sort_values(['Country', 'ConvertedCompYearly'], ascending=[True, False], inplace=True)
```

```python
df[['Country', 'ConvertedCompYearly']].head(50)
```

There are a couple of outlaiers in our data, we'll get to explore that more in the next lesson but for now we want to see how can we locate the largest and smallest values. Because maybe you're sorting to grab the largest or smallest values in a dataframe. If that's the case there is a much more simpler way of doing so.

Let's say we want to see teh 10 higest saliries in our servey. To do that, we can simply use the `nlargest` method. We'll do that first with a series and grab the first 10 

```python
df['ConvertedCompYearly'].nlargest(10)
```

So now it did give us the highest 10 but what if I wanted to see the rest of the survey results. To do that I just need to run it on the dataframe itself 

```python
df.nlargest(10, 'ConvertedCompYearly')
```

If we waned the to see the smallest values instead of the largest then we just use `nsmallest` as such

```python
df.nsmallest(10, 'ConvertedCompYearly')
```
