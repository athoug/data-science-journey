### Grouping Data
Now this topic might be a little confusing, but we'll start slow to get to understand the basics and slowly build up. So firstly, if we want to see results based on `country` or a specific column, then we have to group on that specific column. We can do that with teh `groupby` method. 

Now what does it mean when we say, we're going to group by using the `groupby` function? When we look at the [panda docs](https://pandas.pydata.org/docs/user_guide/groupby.html) it says that the operation includes some kind of combination of 
1. splittling the object
1. applying a function
1. combining the rsults

So we'll walk through each of the processes one at a time so we can see exactly how does that work.
First just for refrence, let's display the values count for each country so that we can see the country that has teh most results for this particular survey

```python
df['Country'].value_counts()
```

Now let's see how to use the `groupby` method on the country column. NOe remember first we will
1. split the object
1. apply the function
1. combine the results

#### Splitting object
first we'll look at slitting the object, and in this case we want to group all the results by country. To do this we simply do the following

```python
df.groupby(['Country'])
```

When running this we see that what we get back is a dataframe group by object `DataFrameGroupBy`, Now what is this object, and what can we do with it?

So this object containes a bunch of groups and to better understand what it is, let's take a look at an individual group that this datframe has. Lets set the `groupby` we just did into a variable so we don't have to constantly type it. 

```python
country_grp = df.groupby(['Country'])
```

Since it grouped by country, then we can access a specific group my country name. So let's grab the group associated with the US, and to do that we need to use the `.get_group` method and passing the name of the group and in this case it's the `United States of America`

```python
country_grp.get_group('Saudi Arabia')
```

__Note__ that this is similar to running filter on our datafame (just as we did a few lessons ago) let's see how we'd do it with a filter 

```python
filt = df['Country'] == 'Saudi Arabia'
df.loc[filt]
```

So now let's go back to our example, we want to see `gender` broken by `country` we can do that with our filtered value by selecting the column and applying `value_counts` on it as such

```python
filt = df['Country'] == 'India'
df.loc[filt]['Gender'].value_counts()
```

Now those are the results of a single country, but if we happen to run it on our dataframe groupby object then it will give us the results of all those countries groups. 
So we use our variable and select the column we want and apply the method on it

```python
country_grp['Gender'].value_counts().head(50)
```

So we can see that we have a series here and the series has multiple indexes (one for country and the other for geneder) so if we want to access the data only for Indeia what we do is simpley attach a `.loc` in the end as follows. Remember if you prefer to see percentages instead of raw values all you have to do is set the `normalize` value to `True`

```python
country_grp['Gender'].value_counts(normalize=True).loc['India']
```

Now we can use it also in results that resumble our aggregate functions such as `mean`, `median` and `mode` So before we looked at the median for salary, but now let's break it down to each country 


```python
country_grp['ConvertedCompYearly'].median().tail(50)
```

Now say I want to see only the result of Germany then what I can do is used the `loc` indexer

```python
country_grp['ConvertedCompYearly'].median().loc['Germany']
```

Now what if we want to see the mean along with the median? then what we can do is use the `agg` method and pass all the aggregate functions we want to run

```python
country_grp['ConvertedCompYearly'].agg(['median', 'mean'])
```

Now as we said before, if we want to narrow it down to a specific country, we can do that by simply by grabing the results by each country name so if we wanted to look at the `mean` and `median` for Saudi Arabia then 

```python
country_grp['ConvertedCompYearly'].agg(['median', 'mean']).loc['Saudi Arabia']
```
