We'll first look at `CSV` since that's what we've been working with through out this series. We've seen reading in data. Now though, let's explore how to `write` back data to a csv

let's take an example of filtering out our data say to a specific country and save the result to do further analysis.
so let's make a filter for servey results from `Saudi Arabia`

```python
filt = df['Country'] == 'Saudi Arabia'
sa_df = df.loc[filt]
sa_df
```

Now say we want to take this filtered dataframe and turn it to a `.csv` file what we do is use the `to_csv` method and pass in the path to where you want the file to be writen 

```python
sa_df.to_csv('./data/modified.csv')
```

## TSV 
another format that we might run into is a tab limited file. it's similar to csv, but instead of your data being seperated by a comma `,` the data instead is seperated by tabs. 

To do this in pandas we'll still use teh same csv methods we've already seen, but we'll pass in a custom seperator. 

to write a file, all we have to do is change the extention to `.tsv` and pass ina `sep` argument specifying that its should be tab `\t`

```python
sa_df.to_csv('./data/modified.tsv', sep='\t')
```

now to read `tsv` we do the same as we did with `csv` but also attach the the `sep='\t'` argument as follows

```python
df_tsv = pd.read_csv('./data/modified.tsv', sep='\t')
df_tsv
```

## JSON
now lets come to the one I work with the most (since I am a javascript developer)
What we'll do is first write our fitered data to a json file. To do that, we use teh `to_json` method however this is a bit different since there are different oreantations we can use with json. 

Let's see what we get with the default arguments, then look into changing it up a bit

```python
sa_df.to_json('./data/modified.json')
```

Once we look at teh data saved, we see that it's stored is a dictionary format. There are other ways to store it we'll not go over all of them but say we want our values to be stored as a list rather than a dictionary. To do that, we can change the `orient` argument 
__note__ if we pass in `records` for `orient` and `lines=True`(which will make each entry into a new line) then it will make it list like

If you want to know more about what we can pass in, just a quick google of `to_json` will give you, all you need,

```python
sa_df.to_json('./data/modified.json', orient='records', lines='True')
```

It's time to start eploring reaing json data. __note__ if you're reading in json and you changed the way you store it (like we did up top) then make sure to specify that when you read in the data as well.


```python
test = pd.read_json('./data/modified.json', orient='records', lines='True')
```
