### Dataframes
lets first learn what a dataframe is and learn how to think about it in terms of a python object. If we look up the definition online, it would say it's a 'two-dimentional' list which basically menas columns and rows. 
We saw it earlier when we loaded the data which was rows and columns so basically, a table, but what is it in native python?

So if we're only using python and not pandas to store this data of rows and columns then how would we do this? We might think of a dictionary to be a good way to store this. Let's take an example.

```python
person = {
    "first": 'Athoug',
    "last": 'Alsoughayer',
    "email": 'athoug@somemail.com',
}
```

```python
people = {
    "first": ['Athoug', 'jane', 'John'],
    "last": ['Alsoughayer', 'Doe', 'Doe'],
    "email": ['athoug@somemail.com', 'janedoe@email.com', 'johndoe@email.com'],
}
```

So as we know, a dictionary is a set of key value pairs. for a single item it would work by having a single value, but if we want to replicate or give the illution of rows an columns, one way to think about it is as key and a list or values as seen with the example above.

So if I want to see the list all I have to do is access them using the key. so lets say, I want to access the email list then 

```python
people['email']
```

So data frames are similar to what we just demonstrated but with more functionality. Actually we can create a dataframe from this dictionary and see what it looks like

```python
people_df = pd.DataFrame(people)
```

```python
people_df
```

we can see now that we created a dataframe, our dictionary looks like a table with rows and columsn. 
Now note we have numbers that don't have a column name, and those are indexes which are basicaly a unique value for our rows (it doesn't need to be unique but we'll talk more about it in the indexes section). 

Now that we have an idea of how to think about a dataframe, it's time to explore how to access information with dataframes

Lets start off by accessing a single column. We can access it the way we did a dictionary as such 

```python
people_df['email'] #  bracket notation (named column/ key)
```

```python
people_df.email # dot notation
```

__note__ that when you use dot notation to access series you might run into errors if the key is similar to an attribute already assigned to the dataframe for example `.count` so stay on the safe side and use bracket notaion

we can see that it doesn't look like the dictonary value returned and that's because what the dataframe gives us is a `series` we can check that using the `type` method 

```python
type(people_df['email']) 
```

### Series
So what is a series? it's basically a list of data, but just like dataframes, it has a lot more functionality than just a normal list. if you look up the definition online it would say it's a 'one-dimentional' array which basically means it's rows of data

So again we can think of a __dataframe__ as rows and columns 
and think of a __series__ as rows of a single column

A dataframe is a contianer of multiple of these series objects

We said that dataframes and series have more functionality so lets explore that. Say we wanted to access multiple columns. to do that we can use the bracket notation and pass a list of the columns that we want.
So if we want the first and last name columns what we can do is 

```python
people_df[['first', 'last']]
```

__note__ now that we got multiple columns, this is no longer a series because remember it (series) are rows of a single column. so when we get multiple columns like the example above, it's just returning another dataframe. And in this case, a filtered down dataframe with just those specific columns.

we can also `slice` it down just as we do with lists (we'll expplore later in a bit)

Now if we want to see a list of columns we can simply use the `.columns` attribute as such

```python
df.columns
```

Now we saw how we can get our columns, now what about our rows? In order to get rows, we can use the `.loc` and `iloc` methods. Let's explore them.

`.iloc[]` allows us to access rows by integer location, hene the name. So if I want to access the first row then we simple `df.iloc[4]` now say we want to acceess the first and third row, then `df.iloc[[0,2]]

```python
df.iloc[4]
```

```python
df.iloc[[0,2]]
```

Now we can also access specific columns so if we thing of `loc` and `iloc` as functions, then the first argument are the rows and teh second is the columns. 

Remember with `iloc` we use indexes so we can't add names we have to access teh column indexes as such 
lets take teh people dataframe example since it's smaller

```python
people_df.iloc[[1,2], 2]
```

```python
people_df.iloc[[1,2], [0,1]]
```

Now let's take a look at `loc`. Now with `.iloc` we were searching by integer location, but with `.loc` we're searching by label. We'll explore the use case when we cover indexes but for now it will look just like `.iloc`

Let's look at our people dataframe one more time

```python
people_df
```

If we look at the numbers in the far left those are our lables. so if I want the first row, it has teh label of zero, so we can do the following

```python
people_df.loc[0]
```

Just with `.iloc` we can get multiple rows as follows

```python
people_df.loc[[1,2]]
```

Just as we did with `iloc` we can specify specific columns for those rows as well, but note that with `loc` we use lables

```python
people_df.loc[[1,2], ['first','last']]
```

## Indexes
we learned earlire how to use default indexes, but now it's time to learn all about custom indexes along with the benifits of doing it.

We'll start off by looking at our peoples dataframe sine it's a smaler dataset which will help us in understanding. Let's have a look at the peoples df

```python
people_df
```

As we can see, we have a nameless numbered columns which are our indexes. Those ar ethe defaults. However, sometimes it would be benificial to have a different identifier for that row, and that's called a `label`. They're usually uniqe but pandas doesn't enforce that.

So looking at the data above, what looks like a good idea for uniqe values. Email seems ligitimate. So if I wanted to view all the email adress

```python
people_df['email']
```

So what if we wanted to set the `email` as the default index, we do that by using `set_index('column name')` 

```python
people_df.set_index('email')
```

```python
people_df
```

Now that is odd, we just set the email to be the index, but when we looked at the data frame again, it shifted back to the default index. The reason is that Pandas doesn't do a lot of these changes in place unless we tell it to do so. That's good because we can experament without worrying about it affecting anything.

So we can change it and have it carry over by adding another argument `inplace=True` right after the set as follows

```python
people_df.set_index('email', inplace=True)
```

```python
people_df
```

Now we can see that it's official the email now is the main index! We can also look specifically at that index using the `.index` property as follows

```python
people_df.index
```

Now the question is, why would it be usuful? Well it's much easier to filter with `.loc` so before we just used the numbered indecese which was more difficult to filter unless you wanted to view specific id's however what if you wanted to find say specific `emails` or another label. In this case custom labels are much more convenient and easier when it comes to filtering

```python
people_df.loc['athoug@somemail.com']
```

Now say I just want that rows last name column, as we mentioned earlier `.loc` takes in the first list for rows, and the second for columns so what we can do to get the name is as such

```python
people_df.loc['athoug@somemail.com', ['first', 'last']]
```

__Note__ that we can't use integer indexes with `.loc` anymore because it set the index to email, but if we still wanted to use integer indexes we can still use `iloc` as follows

```python
people_df.iloc[1]
```

Now say you wanted to revert back to the original default index you can do that by reseting the index with `reset_index` method and of course don't forget to include the `inplace=True` argument to make sure the changers carry over

```python
people_df.reset_index(inplace=True)
```

```python
people_df
```
