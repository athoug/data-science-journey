## Practicing filters

Lets start off by creating teh smaller set of people

```python
people = {
    'first': ['athoug', 'jane', 'john', 'sam', 'zack', 'olivia'],
    'last': ['alsoughayer', 'doe', 'doe', 'simples', 'something', 'mendez'],
    'email':['athoug@email.com', 'jane@email.com', 'john@email.com', 'sam@email.com', 'zack@email.com', 'olivia@email.com'],
    'year': [1990, 1978, 1981, 1997, 1994, 1999],
    'hobby': ['making', 'cooking', 'sports', 'crafts', 'music', 'drama']
}
```

```python
people_df = pd.DataFrame(people)
people_df
```

Lets filter it to people with the last name of `doe`

```python
people_df['last'] == 'doe'
```

We might have assumed we'd get a filtered list with just teh results that matched the comparison but that's not what we got. We got a series with a boolean that states if it matched or not. So basically we got a mask of the data set. Now we want to apply that filter to the dataframe and not the series (which is the list for a single column)

```python
filt = (people_df['last'] == 'doe')
people_df[filt]
```

Now we can also use the `.loc` indexer as well 

```python
people_df.loc[filt]
```

The instructor perfers the `.loc` indexer because he can also just get the columns he wants as follows 

```python
people_df.loc[people_df['last']=='doe', 'email']
```

Lets move on to exploring more filtering options! We'll start off with `&` and `|` operators (__Note__ we can't use the `and` and `or` keywords in python to filter dataframes)

Let's take an example where we want to filter the dataset for a last name of `doe` and a first name of `john`

```python
people_df[(people_df['last']=='doe') & (people_df['first']=='john')]
```

Using `.loc` this time while only getting the `email`

```python
people_df.loc[(people_df['last']=='doe') & (people_df['first']=='john'), 'email']
```

Now lets explore the or `|` operator. Lets say we wnat data where the last name is `alsoughayer` and first name is `zack` 

```python
filt = (people_df['last'] == 'alsoughayer') | (people_df['first'] == 'john')
people_df[filt]
```

```python
people_df.loc[filt, 'email']
```

Now say we want to get the complete oposite of a filter what we can use is that same filter but attach a tilde `~` at the begining 

```python
people_df.loc[~filt, 'email']
```

```python
people_df[~filt]
```

## Practicing on Stack overflow data
it's time to take the new gained filtering knowledge and put it into test with our larger dataset

Let's say we want to look for people who make a salary of a specific amount (70,000) and we only want to see the `country` , '`computer languages` and `salary` columns

```python
schema_df.loc['CompTotal', 'question']
```

```python
filt = (df['CompTotal'] >= 70000)
```

```python
df.loc[filt, ['Country', 'LanguageHaveWorkedWith', 'CompTotal']]
```

Now what we have here is cool, but what if I just wanted to see the result of people from a specific set of countries. We do that by first making a list of the countries we want to filter by

```python
countries = ['United States of America', 'India', 'Canada', 'Germany', 'United Kingdom']
```

Then we can use the filter `isin()` to filter out the rows

```python
filt = (df['Country'].isin(countries)) & (df['CompTotal'] > 70000)
```

```python
df.loc[filt]
```

Now we also have sring filtes. Now lets say we want to see results of people who answered that they use python as a language  

```python
df['LanguageHaveWorkedWith']
```

So we can see that the string has multiple values seperated by a semicolun so we need to use string methods that would help us check if `Python` is in there using the `.srt.contains()` method (__note__ that the `na=False` option is to clear out the `NaN` in our dataset fpr it not to fail)

```python
filt = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
df.loc[filt, 'LanguageHaveWorkedWith']
```
